```python
#Find maximum element in an array
def find_maximum(arr):
    max_element=arr[0]
    for i in arr:
        if i>max_element:
            max_element=i
    return max_element  
print("Maximum element is:",find_maximum(arr))      
```

```python
#Sum of all elements in an array
def sumo(arr):
    total=0
    for i in arr:
      total+=i
    return total

print(sumo(arr))  
      
```

```python
def frequency(arr):
    freq = {}
    for x in arr:
        freq[x] = freq.get(x, 0) + 1
    return freq
print(frequency(arr))
```

```python
def large(arr):
    first=0
    second=0
    
    for i in arr:
        if i>first:
            first=i

    for i in arr:
        if i>second and first!=i:
            second=i        
    return second 
print(large(arr)) 


```

```python
#Linear Search:
arr=[1,2,3,4,5,6]
def linear_Search(arr,key):
    for i in range(0,len(arr)):
        if i==key:
            return i
    return -1
print(linear_Search(arr,4))    
```

```python
#Binary Search 
arr=[1,2,3,4,5,6]
def binarysearch(arr,key):
    low,high=0,len(arr)-1
    while low<=high:
        mid=(low+high)//2
        if arr[mid]==key:
            return mid
        elif arr[mid]<key:
            low=mid+1
        else:    
            high=mid-1
    return  

print(binarysearch(arr,3))    
```

```python
#Occurance of the element:
def occur(arr,key):
    count=0
    for i in arr:
        if i==key:
            count+=1
    return count      
print(occur(arr,2))  
```

```python
#missing element 1 to n
arr=[1,3,4,5,6]
def miss(arr):
    sum1=0
    n=len(arr)
    
    total=(n+1)*(n+2)//2
    
    for i in range(n):
        sum1+=arr[i]
    return total - sum1    
print(miss(arr))
```

```python
#find repeating:
arr=[1,2,3,3,4,5]
def findl(arr):
    seen=set()
    for i in arr:
        if i in seen:
            return i
        seen.add(i)
print(findl(arr))        
        
```

```python
#Find Duplicate Elements:
arr=[1,1,2,3,4,2]
def dup(arr):
    n=len(arr)
    result=[]
    for i in range(n-1):
        if arr[i] in result:
            continue
        for j in range(i+1,n):
            if arr[i]==arr[j]:
                result.append(arr[i])
            
            
    return result
print(dup(arr))      

```

```python
arr=[1,1,2,3,4]
def once(arr):
    result=0
    for i in arr:
        result^=i
    return result
print(once(arr))
```

```python
#Move all zeroes to the end 

arr=[1,2,0,3,4,0,5]
def move(arr):
    result=[]
    count=0
    for i in arr:
        if i==0:
            count+=1
        else:    
            result.append(i)
    result.extend([0]*count)
    return result
print(move(arr))        
```

```python
#Reverse a elements:
arr=[1,2,3,4,5]
def rev(arr):
    l=0
    r=len(arr)-1
    while l < r:
        arr[l], arr[r] = arr[r], arr[l]
        l += 1
        r -= 1
    return arr

print(rev(arr))    
```

```python
def rotate_right(arr, k):
    k = k % len(arr)
    return arr[-k:] + arr[:-k]
print(rotate_right(arr,1))
```

```python
def two_sum(arr, target):
    seen = set()
    for x in arr:
        if target - x in seen:
            return True
        seen.add(x)
    return False
print(two_sum(arr,4))
```

```python
#SUBARRAY max sum of subarray
arr=[1,2,3,4,2,3]
def maxarr(arr,k):
    window=sum(arr[:k])
    sumw=window

    for i in range(k,len(arr)):
        window += arr[i]-arr[i-k]
        sumw=max(sumw,window)
    return sumw
print(maxarr(arr,3))    
```

```python
#Min sum of an subarray
arr=[1,3,4,5,7,2,4]
def minof(arr,k):
    window=sum(arr[:k])
    minsum=window
    for i in range(k,len(arr)):
        window+=arr[i]-arr[i-k]
        minsum=min(minsum,window)
    return minsum
print(minof(arr,3))    
```

```python
#Count subarray of specific number:
def count(arr,k):
    return len(arr)-k+1
print(count(arr,3))
```

```python
#max of all sub array:
def max_of_subarrays(arr, k):
    result = []

    for i in range(len(arr) - k + 1):
        result.append(max(arr[i:i+k]))

    return result
print(max_of_subarrays(arr,5))
```

```python
#Longest SUBARRAY:
def longest(arr,k):
    curr=0
    left=0
    max_len=0

    for i in range(len(arr)):
        curr+=arr[i]

        while curr>k:
            curr-=arr[left]
            left+=1

        
        if i-left+1>max_len:
            max_len=max(max_len,i-left+1)  
            start=left

    return arr[start:start + max_len]
print(longest(arr,4))    
```
