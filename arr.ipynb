```python
arr=[3,7,2,9,4]
```

```python
#Find maximum element in an array
def find_maximum(arr):
    max_element=arr[0]
    for i in arr:
        if i>max_element:
            max_element=i
    return max_element  
print("Maximum element is:",find_maximum(arr))      
```

```python
#Sum of all elements in an array
def sumo(arr):
    total=0
    for i in arr:
      total+=i
    return total

print(sumo(arr))  
      
```

```python
def frequency(arr):
    freq = {}
    for x in arr:
        freq[x] = freq.get(x, 0) + 1
    return freq
print(frequency(arr))
```

```python
def sorte(arr):
    for i in arr:
        if arr[i]>arr[i+1]:
            return False
    return True    
```
