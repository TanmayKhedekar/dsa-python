```python
#Reverse String:
def reverse_string(s):
    return s[::-1]
print(reverse_string(s))
```

```python
#Palindrome:
def is_palindrome(s):
    return s == s[::-1]
print(is_palindrome(s))
```

```python
#Frequency of character
def frequ(s):
    freq={}
    for c in s:
        freq[c]=freq.get(c,0)+1
    return freq
print(frequ(s))    
```

```python
#Anagram = when two sorted string are equal
s='Tanmay'
l='yamnat'
def ana(s,l):
    return sorted(s)==sorted(l)
print(ana(s,l))
```

```python
s = 'tanmay'

def uniq(s):
    result = []
    freq = {}

    for c in s:
        freq[c] = freq.get(c, 0) + 1

    for c in s:
        if freq[c] == 1:
            result.append(c)

    return result

print(uniq(s))
```

```python
#Longest word in the sentence:
s="Tanmay is a good boy"
def longe(s):
    word=s.split()
    longest=""

    for w in word:
        if len(word)>len(longest):
            longest=w
    return longest
print(longe(s))        
```

```python
#Vowels and consonants:
def count_vowels_consonants(s):
    vowels = "aeiouAEIOU"
    v = c = 0

    for ch in s:
        if ch.isalpha():
            if ch in vowels:
                v += 1
            else:
                c += 1

    return v, c
print(count_vowels_consonants(s))
```

```python
#Calculate the digit in the string:
def only_digits(s):
    return s.isdigit()
print(only_digits(s))
```
