```python
#Frequency of character
def frequ(s):
    freq={}
    for c in s:
        freq[c]=freq.get(c,0)+1
    return freq
print(frequ(s))    
```

```python
s = 'tanmay'

def uniq(s):
    result = []
    freq = {}

    for c in s:
        freq[c] = freq.get(c, 0) + 1

    for c in s:
        if freq[c] == 1:
            result.append(c)

    return result

print(uniq(s))
```

```python
#Longest word in the sentence:
s="Tanmay is a good boy"
def longe(s):
    word=s.split()
    longest=""

    for w in word:
        if len(word)>len(longest):
            longest=w
    return longest
print(longe(s))        
```

```python
#Vowels and consonants:
def count_vowels_consonants(s):
    vowels = "aeiouAEIOU"
    v = c = 0

    for ch in s:
        if ch.isalpha():
            if ch in vowels:
                v += 1
            else:
                c += 1

    return v, c
print(count_vowels_consonants(s))
```

```python
def longest_unique_substring(s):
    seen = set()
    left = 0
    max_len = 0
    start = 0

    for right in range(len(s)):
        while s[right] in seen:
            seen.remove(s[left])
            left += 1
        seen.add(s[right])

        if right - left + 1 > max_len:
            max_len = right - left + 1
            start = left

    return s[start:start + max_len]
```

```python
def count_palindromes(s):
    count = 0

    def expand(l, r):
        nonlocal count
        while l >= 0 and r < len(s) and s[l] == s[r]:
            count += 1
            l -= 1
            r += 1

    for i in range(len(s)):
        expand(i, i)
        expand(i, i+1)

    return count
```

```python
def substring_k_distinct(s, k):
    freq = {}
    left = 0
    result = []

    for right in range(len(s)):
        freq[s[right]] = freq.get(s[right], 0) + 1

        while len(freq) > k:
            freq[s[left]] -= 1
            if freq[s[left]] == 0:
                del freq[s[left]]
            left += 1

        if len(freq) == k:
            result.append(s[left:right+1])

    return result
```
