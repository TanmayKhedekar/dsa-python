```python
class Node:
    def __init__(self, data):   # âœ… correct constructor
        self.data = data
        self.left = None
        self.right = None

# Creating the tree
root = Node(10)
root.left = Node(20)
root.right = Node(30)
root.right.left = Node(25)

```

```python
#Inorder:
def inorder(root):
    if root:
        inorder(root.left)
        print(root.data,end=" ")
        inorder(root.right)

inorder(root)

```

```python
#preorder
def preorder(root):
    if root:
        print(root.data,end=" ")
        preorder(root.left)
        preorder(root.right)
preorder(root)        
```

```python
#PostOrder:
def postorder(root):
    if root:
        postorder(root.left)
        postorder(root.right)
        print(root.data,end=" ")
postorder(root)        
```

```python
#level order Traversal BFS:
from collections import deque

def level_order(root):
    if not root:
        return

    q = deque([root])

    while q:
        node = q.popleft()
        print(node.data, end=" ")

        if node.left:
            q.append(node.left)
        if node.right:
            q.append(node.right)
level_order(root)            
```

```python
#Height of the tree:

def diameter(root):
    max_diameter = 0

    def height(node):
        nonlocal max_diameter
        if node is None:
            return 0

        left = height(node.left)
        right = height(node.right)

        max_diameter = max(max_diameter, left + right + 1)
        return 1 + max(left, right)

    height(root)
    return max_diameter
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None
```

```python
def insert_bst(root, val):
    if root is None:
        return Node(val)

    if val < root.data:
        root.left = insert_bst(root.left, val)
    else:
        root.right = insert_bst(root.right, val)

    return root
```

```python
def search_bst(root, key):
    if root is None or root.data == key:
        return root

    if key < root.data:
        return search_bst(root.left, key)
    else:
        return search_bst(root.right, key)
```

```python
#Traversal in bst:
def inorder(root):
    if root:
        inorder(root.left)
        print(root.data, end=" ")
        inorder(root.right)
```

```python
#min and max in BST
def find_min(root):
    while root.left:
        root = root.left
    return root.data

def find_max(root):
    while root.right:
        root = root.right
    return root.data
```

```python
#Delete node
def delete_bst(root, key):
    if root is None:
        return root

    if key < root.data:
        root.left = delete_bst(root.left, key)
    elif key > root.data:
        root.right = delete_bst(root.right, key)
    else:
        # Case 1 & 2
        if root.left is None:
            return root.right
        elif root.right is None:
            return root.left

        # Case 3: inorder successor
        temp = root.right
        while temp.left:
            temp = temp.left

        root.data = temp.data
        root.right = delete_bst(root.right, temp.data)

    return root
```

```python
#Mirror of the tree:
def mirror(root):
    if root is None:
        return None

    root.left, root.right = root.right, root.left
    mirror(root.left)
    mirror(root.right)

    return root

```

```python
#Check two trees are same:
def is_identical(t1, t2):
    if t1 is None and t2 is None:
        return True
    if t1 is None or t2 is None:
        return False

    return (t1.data == t2.data and
            is_identical(t1.left, t2.left) and
            is_identical(t1.right, t2.right))
```
