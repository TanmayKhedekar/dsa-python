```python
class Node:
    def __init__(self, data):   # âœ… correct constructor
        self.data = data
        self.left = None
        self.right = None

# Creating the tree
root = Node(10)
root.left = Node(20)
root.right = Node(30)
root.right.left = Node(25)

```

```python
#Inorder:
def inorder(root):
    if root:
        inorder(root.left)
        print(root.data,end=" ")
        inorder(root.right)

inorder(root)

```

```python
#preorder
def preorder(root):
    if root:
        print(root.data,end=" ")
        preorder(root.left)
        preorder(root.right)
preorder(root)        
```

```python
#PostOrder:
def postorder(root):
    if root:
        postorder(root.left)
        postorder(root.right)
        print(root.data,end=" ")
postorder(root)        
```

```python
#level order Traversal BFS:
from collections import deque

def level_order(root):
    if not root:
        return

    q = deque([root])

    while q:
        node = q.popleft()
        print(node.data, end=" ")

        if node.left:
            q.append(node.left)
        if node.right:
            q.append(node.right)
level_order(root)            
```
