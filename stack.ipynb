```python
def push(stack,x):
    return stack.append(x)

def pop(stack):
    if len(stack)==0:
        print("Stack is empty")
    return stack.pop()

def peek(stack):
    if len(stack)==0:
        return None
    return stack[-1]

push(stack,10)
push(stack,20)
push(stack,30)
push(stack,40)

pop(stack)
peek(stack)
```

```python
#reverse string using stack:
def reverse(s):
    stack=[]
    for ch in s:
        stack.append(ch)

    result=""
    while stack:
        result+=stack.pop()

    return result
print(reverse("tanmay"))        

```

```python
#Valid perenthesis:
def valid(s):
    stack=[]
    pair={'(':')','{':'}','[':']'}
    for c in s:
        if c in pair.values():
            stack.append(c)
        else:
            if not stack or stack.pop()!=pair[c]:
                return False
    return len(stack)==0
print(valid('{[()]}'))        
        
```

```python
#Stack using linked:
class StackNode:
    def __init__(self, data):
        self.data = data
        self.next = None

class Stack:
    def __init__(self):
        self.top = None

    def push(self, x):
        node = StackNode(x)
        node.next = self.top
        self.top = node

    def pop(self):
        if not self.top:
            return None
        x = self.top.data
        self.top = self.top.next
        return x
```

```python
def postfix_eval(expr):
    stack = []

    for ch in expr:
        if ch.isdigit():
            stack.append(int(ch))
        else:
            b = stack.pop()
            a = stack.pop()

            if ch == '+':
                stack.append(a + b)
            elif ch == '-':
                stack.append(a - b)
            elif ch == '*':
                stack.append(a * b)
            elif ch == '/':
                stack.append(a // b)

    return stack[0]
```

```python
#infix to postfix:
def infix_to_postfix(expr):
    prec = {'+':1, '-':1, '*':2, '/':2}
    stack = []
    output = ""

    for ch in expr:
        if ch.isalnum():
            output += ch
        elif ch == '(':
            stack.append(ch)
        elif ch == ')':
            while stack and stack[-1] != '(':
                output += stack.pop()
            stack.pop()
        else:
            while stack and stack[-1] != '(' and prec[stack[-1]] >= prec[ch]:
                output += stack.pop()
            stack.append(ch)

    while stack:
        output += stack.pop()

    return output
```
