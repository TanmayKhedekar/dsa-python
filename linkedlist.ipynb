```python
class node:
    def __init__(self,data):
        self.data=data
        self.next=None
        
```

```python
#Creating a linked list:
n1= node(10)
n2=node(20)
n3=node(30)

n1.next=n2
n2.next=n3
head=n1
```

```python
#Traverse Linked list:
def traverse(head):
    temp=head
    while temp:
        print(temp.data, end="->")
        temp=temp.next
    print(None)   

print(traverse(head))     
```

```python
#Now Insert the element at the beginning:
def insertingatbeginning(head,data):
    newnode=node(data)
    newnode.next=head
    return newnode
head=insertingatbeginning(head,5)

print(traverse(head))  
```

```python
#insert at the end :
def attheend(head,data):
    newnode=node(data)

    if not head:
        return newnode
    
    temp=head
    while temp.next:
        temp=temp.next
    temp.next=newnode
    return head
print(attheend(head,40))
print(traverse(head))  
```

```python
#Insert at any position:
def atany(head,data,pos):
    newnode=node(data)

    if pos==0:
        newnode.next=head
        return newnode
    
    temp=head
    for _ in range(pos-1):
        temp=temp.next
    newnode.next=temp.next
    temp.next=newnode

    return head
print(atany(head,25,3))
print(traverse(head))      
    
```

```python
#deletion form the linkedlist:
def delete(head,val):
    if not head:
        return None
    
    if head.next == val:
        return head.next
    
    temp=head
    while temp.next and temp.next.data!=val:
        temp=temp.next

    if temp.next:
        temp.next=temp.next.next

    return head

print(delete(head,20))
print(traverse(head))          
```

```python
#search
def search(head,key):
    temp=head
    while temp:
        if temp.data==key:

            return True
        temp=temp.next

    
    return head
print(search(head,25))
```

```python
def reverse_recursive(head):
    if not head or not head.next:
        return head

    new_head = reverse_recursive(head.next)
    head.next.next = head
    head.next = None

    return new_head
```

```python
def reverse_list(head):
    prev = None
    curr = head

    while curr:
        next_node = curr.next   # store next
        curr.next = prev        # reverse link
        prev = curr             # move prev
        curr = next_node        # move curr

    return prev   # new head
```

```python
#Middle element
def find_middle(head):
    slow = fast = head

    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next

    return slow.data
```

```python
#Floyd's Cycle detection:
def cycle(head):
    slow=fast=head
    while fast and fast.next:
        slow=slow.next
        fast=fast.next.next
        if slow==fast:
            return True
    return False
print(cycle(head))
    
```

```python
#Nth node from the end:
def nth_from_end(head, n):
    slow = fast = head

    for _ in range(n):
        if not fast:
            return None
        fast = fast.next

    while fast:
        slow = slow.next
        fast = fast.next

    return slow.data
print(nth_from_end(head,25))
```

```python
#Merging two linked list
def merge_lists(l1, l2):
    dummy = node(0)
    curr = dummy

    while l1 and l2:
        if l1.data < l2.data:
            curr.next = l1
            l1 = l1.next
        else:
            curr.next = l2
            l2 = l2.next
        curr = curr.next

    curr.next = l1 if l1 else l2
    return dummy.next
```

```python
#Remove duplicated:
def remove_duplicates(head):
    curr = head

    while curr and curr.next:
        if curr.data == curr.next.data:
            curr.next = curr.next.next
        else:
            curr = curr.next

    return head

```

```python
#Lengthof the linked list:
def length(head):
    count = 0
    while head:
        count += 1
        head = head.next
    return count
```
